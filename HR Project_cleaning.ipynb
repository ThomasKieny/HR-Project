## Clean our data

```python
#Import our data
df_base = pd.read_csv(r'C:\Users\kieny\Documents\Thomas\Data\PROJETS\01_Projet RH\HRDataset_v14.csv')
df_base.head()
```

```python
# Delete variables we will not be using in this project

delete_var = ['CitizenDesc', 'HispanicLatino','RaceDesc','MaritalStatusID','DeptID']
df = df_base.drop(columns=delete_var)
df.info()
```

```python
#Adjust variables types when necessary:
# 'DOB', 'DateofHire', 'DateofTermination', 'LastPerformanceReview_Date' : in date format
# 'ManagerID' in int

df['DOB'] = pd.to_datetime(df['DOB'])
df['DateofHire'] = pd.to_datetime(df['DateofHire'])
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'])
df['LastPerformanceReview_Date'] = pd.to_datetime(df['LastPerformanceReview_Date'])
```

```python
#Adjust values in 'Sex'
df['Sex'] = df['Sex'].replace('M ', 'M')
```

```python
#Check missing variable for 'ManagerID. We identify for which manager are we missing an ID number (ManagerID). It seems that Webster Butler has no ID attributed.

df[df['ManagerID'].isna()]
```

```python
#We identify missing ID for the manager Webster Butler. We need to check if this manager already has an attributed ID.

df[df['ManagerName']=='Webster Butler']
```

```python
#Webster Butler is attributed with ID 39. We need to replace NaN in the ManagerID column with 39.

df['ManagerID'].fillna(value =39.0, inplace = True)
df['ManagerID'] = df['ManagerID'].astype(int)
```

```python
#We need to check for duplicates

df.duplicated().sum()
```

```python
#There is no duplicate in the dataframe. It looks we have a clean database to work on from now. 
```

```python
#Rename Employee_Name column to have coherence in the naming convention of the dataframe 

df = df.rename(columns = {'Employee_Name':'EmpName'})
```

```python
#Put the EmpName values in the same format as the ManagerName values

df.set_index(pd.Series(df['EmpName'].str.split(',').str[::-1]).apply(lambda x:' '.join(x)), inplace = True)
df.drop(columns = 'EmpName', inplace = True)
df = df.reset_index()
df.head()
```

```python
#We can extract our dataframe

df.to_csv('HRDataset_clean.csv', index = False)
```
